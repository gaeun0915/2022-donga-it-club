```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots()  #figures 와 axes 객체 할당

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1,Y_1) #plot 함수를 사용하여 그래프 생성
ax.set(title ='cos graph', #그래프 제목, x축라벨, y축 라벨 설정
      xlabel='X',
      ylabel='Y');

plt.show()  #그래프 출력
```

```python
fig=plt.figure() # figure 반환
fig.set_size_inches(10,10) #figure의 크기 지정

ax_1=fig.add_subplot(1,2,1) #첫 번째 그래프 생성 행은 1개 열은 2개 
ax_2=fig.add_subplot(1,2,2) #두 번째 그래프 생성

ax_1.plot(X_1,Y_1,c='c')  #첫 번째 그래프 설정
ax_2.plot(X_2,Y_2,c='m')  #두 번째 그래프 설정
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)  # 행은 1개이고 열은 2개인 그림 두개의 축 개체 생성
print(ax)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1,1,100) # x값들과 y_n 값들을 생성한다.
y_1=np.sin(x)
y_2=np.cos(x)
y_3=np.tan(x)
y_4=np.exp(x)

fig, ax = plt.subplots(2,2) #2 x 2 figure 객체를 생성함

ax[0,0].plot(x, y_1)
ax[0,1].plot(x, y_2)
ax[1,0].plot(x, y_3)
ax[1,1].plot(x, y_4)

plt.show()
```

```python
ax1 = plt.subplot(3,2,1)
plt.plot(x, y_1)
ax2 = plt.subplot(3,2,2)
plt.plot(x, y_1)
ax3 = plt.subplot(3,1,2)
plt.plot(x, y_1)
ax4 = plt.subplot(3,2,5)
plt.plot(x, y_1)
ax5 = plt.subplot(3,2,6)
plt.plot(x, y_1)

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value +100 for value in X]

plt.plot(X_1,Y_1, color='bisque')
plt.plot(X_2,Y_2,c='lightpink')

plt.show()
```

```python
plt.plot(X_1,Y_1, color='m', linestyle="dashed")
plt.plot(X_2,Y_2,c='r',ls="dotted")

plt.title('Two lines')
plt.show()
```

```python
fig=plt.figure()
fig.set_size_inches(10,10)

ax_1=fig.add_subplot(1,2,1)
ax_2=fig.add_subplot(1,2,2)

ax_1.plot(X_1,Y_1,c='coral')
ax_1.set_title('Figure 1')
ax_2.plot(X_2,Y_2,c='lightskyblue')
ax_2.set_title('Figure 2')
plt.show()
```

```python
plt.plot(X_1,Y_1, color='m', linestyle="dashed",label='line_1')
plt.plot(X_2,Y_2,c='r',ls="dotted",label='line_2')

plt.legend(shadow=True, fancybox=False, loc='upper right')

plt.title('$y=ax+b$')
plt.xlabel('$x_lines$')
plt.ylabel('y_line')
```
